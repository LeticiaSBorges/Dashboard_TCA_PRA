Para filtar apenas o quantitativo de imóveis com licenciamento, foi criado um dataframe selecionando apenas a linhas com valor igual a 1 na coluna "N_LICENCIAMENTOS", posteriormente foi excluida essa coluna e exportada para excel.

Para saber o quantitativos de TCAs é necessário fazer a contagem de codigo de imóveis por região de integração, contudo aqui realizado a contagem de valores 1 e 0 de N_LICENCIAMENTOS por região de integração. Este dataframe foi criado, exportado em excel e chamado no algoritmo.

Logo em seguida foi criado um vetor com o nome das regiões de integração e organizados em ordem alfabetica.

Após a solicitação dos dados exportado em excel, foi selecioado apenas o valores igual a 1 na coluna N_LICENCIAMENTOS, excluida as colunas REGIAO_INTEGRACAO e N_LICENCIAMENTOS, acrescentao o vetor de região de integração criado anteriormente, renomeado as colunas para  'TCA' e 'Região de Integração', e por fim organizado as colunas do dataframe em 'Região de Integração' e 'TCA'.

```python
tabTCA = dados.groupby(by=['REGIAO_INTEGRACAO','N_LICENCIAMENTOS']).sum().groupby(level=[0]).cumsum()
#print(tabTCA)

vetorRegiao_Int = dados['REGIAO_INTEGRACAO'].unique()
vetorRegiao_Int = vetorRegiao_Int.flatten().tolist()
vetorRegiao_Int = sorted(vetorRegiao_Int)

tabTCA.to_excel (r'C:\Users\letic\Documents\Python Scripts\PRA TCA\TCA_Regiao_Int.xlsx', index = True, header=True)

tca = pd.read_excel('TCA_Regiao_Int.xlsx')
#print(tca)
RegInt_TCA = tca.loc[tca['N_LICENCIAMENTOS']==1]
RegInt_TCA.drop(['REGIAO_INTEGRACAO','N_LICENCIAMENTOS'], axis=1, inplace=True)
RegInt_TCA['Região de Integração'] = vetorRegiao_Int
RegInt_TCA.set_axis(['TCA', 'Região de Integração'], axis='columns', inplace=True)
RegInt_TCA = RegInt_TCA[['Região de Integração','TCA']]
RegInt_TCA.to_excel (r'C:\Users\letic\Documents\Python Scripts\PRA TCA\TCA.xlsx', index = False, header=True)
print(RegInt_TCA)
```

#### Trabalhando com os outros dados da tabela

A primeira maneia de agrupar todos os valores das variáveis "AREA_IMOVEL", "AREA_RECOMPOR-APP", "AREA_RECOMPOR-RL" e "N_LICENCIAMENTOS", e utilizando a função *pivot_table*, como pode ser observado abaixo:

```python
Area_Imoveis_df = pd.pivot_table(data_df, index=["REGIAO_INTEGRACAO"], values=['N_LICENCIAMENTOS',
                                                                               'AREA_IMOVEL','AREA_RECOMPOR-RL', 
                                                                               'AREA_RECOMPOR-APP'], aggfunc=[np.sum])
display(Area_Imoveis_df)
```

Contudo, foi utilizado a função *groupby().sum()*. Em seguida foi excluida a coluna "AUR_Consolidada" por não utilizada no relatório e a coluna "N_LICENCIAMENTOS", pois já foi criada anteriomente (podendo ser utilizada essa também).

```python
data_df2 = data_df.groupby('REGIAO_INTEGRACAO').sum()
data_df2.drop(['AUR_Consolidada', 'N_LICENCIAMENTOS', 'Ano_Termo_Compromisso', 'Mes_Termo_Compromisso'], axis=1, inplace=True)
display(data_df2)
```

Os passos para organização dos dados foram:

* Criação de um index de 1 ao tamanho da tabela dinamica por região de integração;

* Acrescentar o vetor com nome dos municípios e um vetor com o index, criado no etapa anterior;

* Renomear a coluna do dataframe;

* Organizar as colunas.

```python
#Criação de um index de 1 ao tamanho da tabela dinamica por região de integração
x = [0]*len(data_df2)
j = 1
for i in range(len(data_df2)):
    x[i] = j + i

#regInt = ['Araguai', 'Baixo Amazônas', 'Carajás', 'Guamá', 'Lago de Tucuruí',
#         'Marajó', 'Rio Capim', 'Tapajós', 'Tocantis', 'Xingu']

#Acrescentando o vetor com nome dos municípios e um vetor com o index, criado acima
new_df=data_df2.assign(Regiao_Integracao = vetorRegiao_Int, Index = x)
new_df.set_index('Index', inplace = True)

#Renomeado o dataframe
new_df.set_axis(['Área do Imóvel','Área a Recompor - RL', 
               'Área a Recompor - APP', 'Região de Integração'], 
              axis='columns', inplace=True)

#tab1.rename(columns={'REGIAO_INTEGRACAO':'Região de Integração',
#                     'N_LICENCIAMENTOS': 'Número de Licenciamentos', 'AREA_IMOVEL': 'Área do Imóvel',
#                           'AREA_RECOMPOR-RL':'Área a Recompor - RL', 'AREA_RECOMPOR-APP':'Área a Recompor - APP'}, 
#            inplace=True)

#Reorganizado as colunas
new_df = new_df.reindex(columns=['Região de Integração','Área do Imóvel','Área a Recompor - RL','Área a Recompor - APP'])
display(new_df)

new_df.to_excel (r'C:\Users\letic\Documents\Python Scripts\PRA TCA\Dados_TCA.xlsx', index = False, header=True)
```

#### Unindo os dataframes

Para a finalização final da base de dados deve apenas unir os dataframes criados: *N_Imoveis_Licenc*, *TCA* e *Dados_TCA*.

```python
N_Imoveis_Licenc = pd.read_excel('N_Imoveis_Licenc.xlsx')
N_Imoveis_Licenc.set_axis(['Região de Integração','n° de Imovéis com Licencimento'], axis='columns', inplace=True)
#print(N_Imoveis_Licenc)

TCA = pd.read_excel('TCA.xlsx')
TCA.set_axis(['Região de Integração','TCA'], axis='columns', inplace=True)
#print(TCA)

Dados_TCA = pd.read_excel('Dados_TCA.xlsx')
#display(Dados_TCA)

tab_dinamica = N_Imoveis_Licenc.merge(TCA, on = "Região de Integração", how = "left")
tab_dinamica = tab_dinamica.merge(Dados_TCA, on = "Região de Integração", how = "left")
tab_dinamica = tab_dinamica.sort_values(by=['Região de Integração'])
display(tab_dinamica)

tab_dinamica.to_excel (r'C:\Users\letic\Documents\Python Scripts\PRA TCA\Resultados\Geral\tab_dinamica_TCA.xlsx', index = False, header=True)
```
