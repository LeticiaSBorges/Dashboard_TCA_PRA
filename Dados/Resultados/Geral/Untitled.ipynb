```python
import numpy as np
import pandas as pd
df = pd.read_excel("tab_dinamica_TCA.xlsx", engine='openpyxl')

total_n_Lic = np.sum(df['n_de_Imoveis_com_Licencimento'])
total_tca = np.sum(df['TCA'])
total_AreaImovel = round(np.sum(df['AreaImovel']),2)
total_AR_RL = round(np.sum(df['AR_RL']),2)
total_AR_APP = round(np.sum(df['AR_APP']),2)
```

```python
import pandas as pd
pd.options.display.float_format = '{:,.2f}'.format
df = pd.read_excel("tab_dinamica_TCA.xlsx", engine='openpyxl')
df = df.append({'RegInt': 'Total','n_de_Imoveis_com_Licencimento': total_n_Lic, 
               'TCA': total_tca, 'AreaImovel': total_AreaImovel,
               'AR_RL': total_AR_RL, 'AR_APP': total_AR_APP }, ignore_index=True)
display(df)
```
